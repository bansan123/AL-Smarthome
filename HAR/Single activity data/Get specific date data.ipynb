```python
import pandas as pd
```

```python
hh102_s = pd.read_csv("./hh102_single.txt", sep='\t', names=['timestamp','sensor','Translate01','Translate02','Message','SensorType','annotation' ] )
```

```python
hh102_s
```

```python
hh102_s["timestamp"] = pd.to_datetime(hh102_s['timestamp'])
hh102_s["daycase"] = hh102_s["timestamp"].apply(lambda x: str(x.year) + "-" + str(x.month) + "-" + str(x.day))
hh102_s
```

```python
date = "2011-6-20"
data = hh102_s[hh102_s["daycase"]==date]
data["timestamp"]= data['timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S')
data.to_csv("./hh102_"+date+".txt", sep='\t', index=False, header=False)
```
