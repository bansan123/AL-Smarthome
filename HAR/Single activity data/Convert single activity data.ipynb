```python
import pandas as pd
```

```python
hh102 = pd.read_csv("../hh102.ann.txt", sep='\t', names=['timestamp','sensor','Translate01','Translate02','Message','SensorType','annotation' ] )
```

```python
hh102 = hh102.drop(hh102[hh102["annotation"]=="Other_Activity"].index,axis=0)
```

```python
hh102 = hh102.reset_index(drop = True)
```

```python
def find_same_value(series, num_dead_thresh):
    slide_list = [series.index[0]]
    slide_list_all = []
    for i in range(series.index[0],series.index[-1]):
        j = i + 1
        
        if series[j] == series[i]:
            slide_list.append(j)
        else:
            slide_list.clear()
            slide_list.append(j)
        # print("slide_list:",slide_list)
        if len(slide_list) >= num_dead_thresh:
            target_list = slide_list.copy()
            slide_list_all.append(target_list)
#     print("slide_list_all:",slide_list_all)
    index= []  # combine the index 
    # keep the longest sequence
    for i in range(len(slide_list_all) - 1):
        if set(slide_list_all[i]) < set(slide_list_all[i + 1]):
            index.append(i)
    m = {i: element for i, element in enumerate(slide_list_all)}
    [m.pop(i) for i in index]
    return list(m.values())
```

```python
series = hh102.iloc[:,-1]
samelist = find_same_value(series, 4)
```

```python
index = []
for i in samelist:
    index.append(i[-1])
```

```python
df_single = hh102.iloc[index]
```

```python
df_single = df_single.reset_index(drop=True)
df_single
```
