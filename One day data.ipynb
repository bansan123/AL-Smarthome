```python
hh102["timestamp"] = pd.to_datetime(hh102['timestamp'], format='%d/%m/%Y %H:%M:%S')
```

```python
hh102["daycase"] = hh102["timestamp"].apply(lambda x: str(x.year) + "-" + str(x.month) + "-" + str(x.day))
```
